## Monte Carlo simulation is a powerful technique for risk prediction and analysis. 

This notebook explains how to use it for risk prediction and provide a real-world use case with a Python implementation.


### The Monte Carlo method for risk prediction involves the following steps:
- Define the model and its parameters
- Specify probability distributions for uncertain variables
- Generate random samples from these distributions
- Run the model multiple times with different samples
- Analyze the distribution of outcomes


### Real-World Use Case: Project Cost Estimation
#### Let's consider a construction project where we want to estimate the total cost and assess the risk of cost overruns.
- 1 Model Parameters:
- 2 Labor costs
- 3 Material costs
- 4 Equipment rental costs
- 5 Unexpected expenses
- 6 Each of these parameters has some uncertainty, which we'll model using probability distributions.

```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_project_cost(num_simulations=10000):
    # Define probability distributions for each cost component
    labor_cost = np.random.normal(50000, 5000, num_simulations)  # Mean: $50,000, Std Dev: $5,000
    material_cost = np.random.triangular(30000, 35000, 45000, num_simulations)  # Min: $30,000, Mode: $35,000, Max: $45,000
    equipment_cost = np.random.uniform(10000, 20000, num_simulations)  # Min: $10,000, Max: $20,000
    unexpected_cost = np.random.exponential(5000, num_simulations)  # Mean: $5,000

    # Calculate total project cost for each simulation
    total_costs = labor_cost + material_cost + equipment_cost + unexpected_cost

    return total_costs
```

```python
# Run the simulation
simulation_results = monte_carlo_project_cost()

# Analyze results
mean_cost = np.mean(simulation_results)
median_cost = np.median(simulation_results)
std_dev = np.std(simulation_results)
percentile_95 = np.percentile(simulation_results, 95)

print(f"Mean project cost: ${mean_cost:.2f}")
print(f"Median project cost: ${median_cost:.2f}")
print(f"Standard deviation: ${std_dev:.2f}")
print(f"95th percentile cost: ${percentile_95:.2f}")

# Visualize results
plt.figure(figsize=(10, 6))
plt.hist(simulation_results, bins=50, edgecolor='black')
plt.title('Distribution of Project Costs')
plt.xlabel('Total Cost ($)')
plt.ylabel('Frequency')
plt.axvline(mean_cost, color='r', linestyle='dashed', linewidth=2, label=f'Mean: ${mean_cost:.2f}')
plt.axvline(percentile_95, color='g', linestyle='dashed', linewidth=2, label=f'95th Percentile: ${percentile_95:.2f}')
plt.legend()
plt.show()
```

### The code explained step by step 

- 1- Defines probability distributions for each cost component:
- 2- Labor cost: Normal distribution
- 3- Material cost: Triangular distribution
- 4- Equipment rental cost: Uniform distribution
- 5- Unexpected expenses: Exponential distribution
- 6- Runs 10,000 simulations, calculating the total cost for each.
- 7- Analyzes the results, calculating mean, median, standard deviation, and 95th percentile of the total costs.
- 8- Visualizes the distribution of project costs using a histogram.
